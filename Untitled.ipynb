```python
import numpy as np 
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers import *
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img
%matplotlib inline
import tensorflow_hub as hub
from tensorflow.keras.applications import mobilenet_v2
from tensorflow.keras.applications.mobilenet_v2 import decode_predictions
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    horizontal_flip=True,
    zoom_range=0.2
)
valid_datagen = ImageDataGenerator(
    rescale=1./255,
 horizontal_flip=True
)

```

```python
train_generator = train_datagen.flow_from_directory('train',target_size=(224,224),batch_size=32)
valid_generator = train_datagen.flow_from_directory('valid',target_size=(224,224),batch_size=32)
```

```python
train_generator.class_indices
```

```python
base_model = mobilenet_v2.MobileNetV2(input_shape=(224,224,3), input_tensor=None,
                                                include_top=False)
base_model.trainable = False
base_model.summary()
```

```python
from tensorflow.keras.callbacks import ReduceLROnPlateau ,ModelCheckpoint,EarlyStopping



earlystop = EarlyStopping(monitor='val_accuracy',patience=6)
learning_rate_reduction = ReduceLROnPlateau(monitor='val_accuracy', 
                                            patience=3, 
                                            verbose=1, 
                                            factor=0.5, 
                                            min_lr=0.00001)
checkpoint = ModelCheckpoint('checkpoints\\model-{epoch:03d}-{accuracy:03f}-{val_accuracy:03f}.h5', verbose=1, monitor='val_accuracy',save_best_only=True, mode='max')  

callbacks = [checkpoint, learning_rate_reduction,earlystop]
```

```python
model=tf.keras.Sequential()
model.add(base_model)
model.add(GlobalAveragePooling2D())
model.add(Dense(256,activation='relu'))
model.add(Dense(4,activation="softmax"))
model.compile(loss=tf.keras.losses.categorical_crossentropy,optimizer=tf.keras.optimizers.Adam(0.001),metrics=['accuracy'])
model.build(input_shape=[None,224,224,3])
history=model.fit(train_generator,epochs=20,validation_data=valid_generator,callbacks=callbacks)
```

```python
model.save('model')
```

```python
plt.figure(figsize=(10,10))
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper right')
plt.show()
# summarize history for loss
plt.figure(figsize=(10,10))

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper right')
plt.show()
```

```python
def get_class(x):
    inv_map = {v: k for k, v in train_generator.class_indices.items()}
    return inv_map[x]
```

```python
x,y=next(valid_generator)
pred=model.predict(x)
title=np.zeros((32,4))
for a in range(32):
    title[a][np.argmax(pred[a])]=1
fig = plt.figure(figsize=(16,16))
for i in range(32):
    ax = fig.add_subplot(6,6,i+1)
    imgplot = plt.imshow(x[i])
    if list(y[i])!=list(title[i]):
        ax.set_title(get_class(np.argmax(y[i]))+'-'+get_class(np.argmax(title[i])),color='r')
    else:
         ax.set_title(get_class(np.argmax(y[i]))+'-'+get_class(np.argmax(title[i])))
            
    plt.xticks([])
    plt.yticks([]);
```
